```python
#importing required libraries

import pandas as pd
from newspaper import Article
import itertools
import csv
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time
```

```python
#making lists of all the required details

headings=[] 
property_price = []
property_price_per_unit = []
property_area=[]
property_facing=[]
Current_status=[]
property_featues=[]
property_posted_by=[]
property_posted_on=[]
```

```python
#extracting the html code snippets

driver = webdriver.Chrome('C:\Program Files (x86)\chromedriver.exe')
driver.get("https://www.propertiesguru.com/residential-search/2bhk-residential_apartment_flat-for-sale-in-new_delhi")
flag = 0
con = {flag:True}
while con[flag] == True:
    time.sleep(5)
    html = driver.page_source
    page = BeautifulSoup(html , features='lxml')
    top = page.find_all("div" , class_="filter-pro-info")
    body1 = page.find_all("div" , class_="row filter-pro-details")
    body2 = page.find_all("div" , class_="col-sm-7")
    bottom = page.find_all("div" , class_="filter-property-owner")
    flag = len(top)
    try:
        if con[flag] == True:
            break
    except:
        con[flag]=True     

    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    
driver.close()
```

```python
#appending required elements in lists
for p in top:
    headings.append(p.find('h1', attrs={'class':'filter-pro-heading'}).text)
    property_price.append(p.find('span', attrs={'class':'price'}).text)
    property_price_per_unit.append(p.find('span', attrs={'class':'price-per-unit'}).text)

for q in body1:
    property_area.append(q.find('div', attrs={'class':'col-4'}).text)
    property_facing.append(q.find('div', attrs={'class':'col-3'}).text)
    Current_status.append(q.find('div', attrs={'class':'col-5'}).text)
        
for r in body2:
    property_featues.append(r.find('ul', attrs={'class':'pro-list'}).text)

for s in bottom:
    property_posted_by.append(s.find('span', attrs={'class':'owner-name'}).text)
    property_posted_on.append(s.find('span', attrs={'class':'owner-post'}).text)
```

```python
#converting into dataframes

data = pd.DataFrame()
data['headings'] = headings
data['property_price'] = property_price
data['property_price_per_unit'] = property_price_per_unit
data['property_area'] = property_area
data['property_facing'] = property_facing
data['Current_status'] = Current_status
data['property_featues'] = property_featues
data['property_posted_by'] = property_posted_by
data['property_posted_on'] = property_posted_on
```

```python
#cleaning unwanted text

for i in range(len(data)):
    data['property_price_per_unit'] = data['property_price_per_unit'][i].replace("@", "")
    data['property_area'] = data['property_area'][i].replace("Area", "")
    data['property_facing'] = data['property_facing'][i].replace("Facing", "")
    data['Current_status'] = data['Current_status'][i].replace("Status", "")
    data['property_featues'] = data['property_featues'][i].replace("\n", "-->")
    data['property_posted_on'] = data['property_posted_on'][i].replace("Posted:", "")
```

```python
#creating csv

data.to_csv('2bhk.csv', index = False)
```
